```python
# Import Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
import numpy as np
import random
```

```python
#read in data
euro = "data/Euro_6_latest.csv"
euro_df = pd.read_csv(euro, encoding='ANSI')
euro_df
```

```python
#delete 1s in the "WLTP CO2 Weighted" column as these only exist for diesel vehicles
#I assume these 1s are an error, given that the unweighted value for these particular (!) vehicles is as expected.

euro_df["WLTP CO2 Weighted"].replace(to_replace=1, value=0)
euro_df
```

```python
#create a new column that tries to take the "WLTP CO2 Weighted" value, and converts NaN values to 0. 

euro_df["Cleaned CO2"] = euro_df["WLTP CO2 Weighted"].fillna(0)
```

```python
#in case of a 0 value, take the value from the "WLTP CO2" column, in case there is a sensible value there.
for index, row in euro_df.iterrows():
    if (row["Cleaned CO2"] == 0):
        euro_df.loc[index, "Cleaned CO2"] = row["WLTP CO2"]
euro_df
```

```python
#fill in blanks on NOx column as 0s
euro_df["THC + NOx Emissions [mg/km]"]=euro_df["THC + NOx Emissions [mg/km]"].fillna(0)

euro_df
```

```python
#NOx + THC is sometimes blank or 0; should be calculated from the two columns
for index, row in euro_df.iterrows():
    if (row["THC + NOx Emissions [mg/km]"] == 0):
        euro_df.loc[index, "THC + NOx Emissions [mg/km]"] = row["Emissions NOx [mg/km]"]+ row["THC Emissions [mg/km]"]
euro_df.loc[euro_df["Manufacturer"]==0]
```

```python
# export the dataframe to a csv for use in projects
euro_df.to_csv("data/Euro_6_CLEAN.csv")
```
