```python
# Import Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
import numpy as np
import random
import seaborn as sns
import requests
import json


pd.set_option('display.max_columns', 50)
```

```python
euro = "data/Euro_6_CLEAN.csv"
euro_df = pd.read_csv(euro, encoding='ANSI')
euro_df
```

# 

```python
euro_emissions = euro_df.groupby(['Manufacturer', 'Model', 'Description'], as_index=False).mean()
euro_emissions
```

```python
manu_emissions = euro_emissions.groupby(['Manufacturer'], as_index=False).mean().sort_values(by='Cleaned CO2')
manu_emissions = manu_emissions[['Manufacturer', 'Cleaned CO2']]
emissions_superstars_pooperstars= manu_emissions["Manufacturer"]


cars = []
euro_emissions_crop = euro_emissions[['Manufacturer', 'Model', 'Description', 'Cleaned CO2']]

for manufacturer in emissions_superstars_pooperstars:
    carDF = euro_emissions_crop.loc[((euro_emissions_crop["Manufacturer"] == manufacturer ))]
    carDF.sort_values(by="Cleaned CO2", ascending=True)
    cars.append(carDF) 
```

```python
# Generate a box plot for each car manufacturer, of the CO2 emissions of their various vehicles.

    
data = []

for car in cars:
    data.append(car["Cleaned CO2"])


fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(manu_emissions["Manufacturer"],manu_emissions["Cleaned CO2"])
ax.set_ylabel('CO2 Emissions')
ax.set_title('Average CO2 of cars by manufacturers')
plt.xticks(manu_emissions["Manufacturer"], rotation='vertical')

fig.set_figheight(8)
fig.set_figwidth(13)

plt.savefig(fname="images/mean emissions by manufacturer.png")


fig, ax = plt.subplots()
ax.set_title('Comparing the emissions distributions of car manufacturers, sorted by mean')
ax.boxplot(data, labels = emissions_superstars_pooperstars, sym="*", widths=(0.8))

plt.xlabel("Manufacturer")
plt.ylabel("WLTP CO2 Weighted")


fig.set_figheight(8)
fig.set_figwidth(13)

plt.savefig(fname="images/boxplots of emissions by manufacturer.png")
```

```python
# show all columns
pd.set_option("display.max_rows", 100)
pd.set_option("display.max_columns", None)
```

```python
emission_date = euro_df.loc[:, ['Powertrain','Cleaned CO2', 'THC + NOx Emissions [mg/km]', 'Date of change']]
emission_date["Date of change"] = pd.to_datetime(emission_date["Date of change"]).dt.strftime('%Y')
emission_date.head()
#emission_date.dtypes
```

```python
emission_rename = emission_date.rename(columns={"Cleaned CO2": "CO2 Emissions [mg/km]", "Date of change": "Year"})
emission_rename.head()
```

```python
emission_groupby = emission_rename.groupby(['Year', 'Powertrain'], as_index=False)
emission_summary = emission_groupby.mean()
emission_summary.head()
```

```python
#ploting a line graph of Average CO2 Emissions by Powertrain 
plt.figure(figsize=(10,10))
sns.factorplot(data=emission_summary, x="Year", y="CO2 Emissions [mg/km]", hue="Powertrain")
plt.title('Average CO2 Emissions by Powertrain')

plt.savefig(fname="images/time series of CO2 emissions by powertrain")
```

```python
#ploting a line graph of Average THC + NOx Emissions by Powertrain 
plt.figure(figsize=(10,10))
sns.factorplot(data=emission_summary, x="Year", y="THC + NOx Emissions [mg/km]", hue="Powertrain")
plt.title('Average THC + NOx Emissions by Powertrain')

plt.savefig(fname="images/time series of THC + NOx emissions by powertrain")
```

# What does this tell us?
For most types of vehicle, CO2 emissions have fallen. HOWEVER, for all types of vehicle other than electric, THC and NOx emissions have risen - perhaps as these are not 'headline' figures, car companies believe people won't notice. NOx emissions are in fact much worse for the enivronment than CO2 (https://www.cbc.ca/news/science/nitrous-oxide-climate-1.5753907)
